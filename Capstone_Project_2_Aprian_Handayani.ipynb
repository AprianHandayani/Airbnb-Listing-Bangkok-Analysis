- Berdasarkan uji normalitas menggunakan Kolmogorov Smirnov diketahui bahwa data tidak terdistribusi normal, sehinggal nilai 0 diisi dengan nilai mediannya

```python
nilai_median = df['price'].median()
df['price'] = df['price'].replace(0, nilai_median)
```

```python
# fungsi untuk mendeteksi outliers
def cek_outliers(kolom):
  Q1 = np.percentile(kolom, 25)
  Q3 = np.percentile(kolom, 75)
  IQR = Q3 - Q1
  lower_range = Q1 - (1.5 * IQR)
  upper_range = Q3 + (1.5 * IQR)

  return lower_range,upper_range
```

```python
lower_bound,upper_bound = cek_outliers(df['price'])

df = df[(df['price'] > lower_bound) & (df['price'] < upper_bound)]
df
```

```python
def label(neighbourhood):
    if neighbourhood in ['Phra Nakhon', 'Dusit', 'Bang Rak', 'Parthum Wan', 'Pom Prap Sattru Phai',
                         'Yan na wa', 'Samphanthawong', 'Phaya Thai', 'Thon buri', 'Bangkok Yai',
                         'Huai Khwang', 'Khlong San', 'Bangkok Noi', 'Bang Phlat', 'Din Daeng',
                         'Sathon', 'Bang Sue', 'Chatu Chak', 'Bang Kho laen', 'Khlong Toei',
                         'Ratchathewi', 'Vadhana']:
        return 'City Center'
    elif neighbourhood in ['Nong Chok', 'Bang Khen', 'Bang Kapi', 'Phra Khanong', 'Min Buri',
                           'Lat Krabang', 'Taling Chan', 'Bang Khun thain', 'Phasi Charoen', 'Nong Khaem',
                           'Rat Burana', 'Bueng Kum', 'Pra Wet', 'Suanluang', 'Chom Thong', 'Don Mueang',
                           'Lat Phrao', 'Bang Khae', 'Lak Si', 'Sai Mai', 'Khan Na Yao', 'Saphan Sung',
                           'Wang Thong Lang', 'Khlong Sam Wa', 'Bang Na', 'Thawi Watthana', 'Thung khru',
                           'Bang Bon']:
        return 'Suburbs'
    else:
        return 'Unknown'
```

# Tipe Kamar



```python
data_pusat = df_pusat_kota['room_type'].value_counts()
data_pusat
```

```python
data_pinggiran = df_pinggiran_kota['room_type'].value_counts()
data_pinggiran
```

```python
plt.figure(figsize=(16, 6))
plt.subplot(1, 2, 1)
plt.pie(data_pusat, labels=data_pusat.index, autopct='%1.1f%%', startangle=210)
plt.title('Proporsi Tipe Kamar di Pusat Kota')
plt.axis('equal')

plt.subplot(1, 2, 2)
plt.pie(data_pinggiran, labels=data_pinggiran.index, autopct='%1.1f%%', startangle=210)
plt.title('Proporsi Tipe Kamar di Pinggiran Kota')
plt.axis('equal')
plt.show()
```

- berdasarkan pie chart diatas, listing dengan tipe kamar Entire home/apt di pusat kota ketersediaannya lebih banyak daripada yang di pinggiran kota
- Hal ini dibuktikan dengan uji proporsi z test yang menunjukkan angka pval sebesar 0.0 yang berartu proporsi tipe kamar Entire home/apt di pusat kota tidak sama dengan di pinggiran kota.

Ho : proporsi tipe kamar Entire home/apt di pinggiran kota sama dengan di pusat kota

Ha : proporsi tipe kamar Entire home/apt di pinggiran kota tidak sama dengan di pusat kota

```python
from statsmodels.stats.proportion import proportions_ztest

count_pinggiran_kota = df_pinggiran_kota['room_type'].value_counts()
count_pusat_kota = df_pusat_kota['room_type'].value_counts()

prop_pinggiran_kota = count_pinggiran_kota / count_pinggiran_kota.sum()
prop_pusat_kota = count_pusat_kota / count_pusat_kota.sum()

count = np.array([count_pinggiran_kota['Entire home/apt'], count_pusat_kota['Entire home/apt']])
nobs = np.array([count_pinggiran_kota.sum(), count_pusat_kota.sum()])

stat, pval = proportions_ztest(count, nobs)

if pval < 0.05:
    print(f'pval sebesar {round(pval, 2)} berarti proporsi tipe kamar Entire home/apt di pinggiran kota tidak sama dengan di pusat kota')
else:
    print(f'pval sebesar {round(pval, 2)} berarti proporsi tipe kamar Entire home/apt di pinggiran kota sama dengan di pusat kota')
```
